### Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program

```python
class Cart():

    def __init__ (self, cart = {}):
        self.cart = cart
        #self.total = 0

    def shop(self):
        item = input("What would you like to get? ").title()
        quantity = int(input(f'How many {item} would you like to get? '))
        price = float(input(f"Price of the {item} is $"))
        cart[item] = [quantity, price]
        #self.total += price*quantity
        #self.cart.update({item:quantity}) # yeni bir dict degil de eski dicte eklenen yenisiyle update ediyor, biraz set bitlestirme gibi
                
    def view(self):
        for item in self.cart:
            print(f'{item.title()}: {cart[item][0]} * ${cart[item][1]}')

    def remove(self):
        while True:
            response = input('\nWhich item would you like to remove? ').title()
            if not response in self.cart:
                print(f'{response} is not in the cart.')
            else:
                del self.cart[response]
                print(f'{response} has been removed from the cart.')
            
            second_response = input("\nWould you like to delete another item? \n'Yes' or 'No'\n").title()
            if second_response == 'No':
                break
              
    def clear(self):
        self.cart.clear()
        
    def check_out(self):
        item_total = []
        for key in cart:
            total = cart[key][0] * cart[key][1] 
            item_total.append(total)
            print(f'{key.title()}: {cart[key][0]} * ${cart[key][1]} = ${total}')
            total_cost = sum(item_total)
            print('-'*10)
            print('Total: $' + str(total_cost))
```
